```python
from abc import ABC,abstractmethod
```

```python

class data_pipeline:
    def __init__(self):
        self.data_souce = None
        self.data_transformations = []
        self.data_destination = None
    
    def add_transformation(self,transform):
        self.data_transformations.append(transform)
        return self
    def select_souce(self,souce):
        self.data_souce = souce
        return self
    def select_destination(self,destination):
        self.data_destination = destination
        return self
        
    def execute(self):
        raw_data = self.data_souce.read_data()
        for transform in self.data_transformations:
            raw_data = transform.transform_data(raw_data)
            print(raw_data)
        print(self.data_destination.write_data(raw_data))
        
class builder_interface(ABC):
    @abstractmethod
    def get_pipeline():
        pass
    @abstractmethod
    def configure_souce(self):
        pass
    
    @abstractmethod
    def configure_transformations(self):
        pass
    
    @abstractmethod
    def configure_destination(self):
        pass
    

class api_reader:
    def read_data(self):
        return "API data read"

class api_write:
    def write_data(self,raw):
        return "API data write"

class local_reader:
    def read_data(self):
        return "local data read"

class local_write:
    def write_data(self,raw):
        return "local data write"
    
class json_reader:
    def read_data(self):
        return "json data read"
    

class json_write:
    def write_data(self,raw):
        return "json data write"


class remove_questione_process:
    def transform_data(self,data):
        return f"process remove question"

class remove_page_process:
    def transform_data(self,data):
        return f"process remove page"
    
```

```python
class json_pipeline_builder(builder_interface):
    def __init__(self,pipeline_process:data_pipeline):
        self.data_pipeline = pipeline_process
        
    def configure_souce(self):
        self.data_pipeline.select_souce(api_reader())
        
    def configure_destination(self):
        self.data_pipeline.select_destination(json_write())
    
    def configure_transformations(self):
        self.data_pipeline.add_transformation(remove_questione_process())
        self.data_pipeline.add_transformation(remove_page_process())
    
    def get_pipeline(self):
        return self.data_pipeline
    
class pdf_pipeline_builder(builder_interface):
    def __init__(self,pipeline_process:data_pipeline):
        self.data_pipeline = pipeline_process
        
    def configure_souce(self):
        self.data_pipeline.select_souce(local_reader())
        
    def configure_destination(self):
        self.data_pipeline.select_destination(api_write())
    
    def configure_transformations(self):
        self.data_pipeline.add_transformation(remove_questione_process())
        self.data_pipeline.add_transformation(remove_page_process())

    def get_pipeline(self):
        return self.data_pipeline
    

if __name__=="__main__":
    raw_data_pipe = data_pipeline()
    pdf_builder = pdf_pipeline_builder(raw_data_pipe)
    pdf_builder.configure_souce()
    pdf_builder.configure_transformations()
    pdf_builder.configure_destination()
    
    process_pipeline = pdf_builder.get_pipeline()
    
    process_pipeline.execute()
```
